```python
##### Co
import matplotlib.pyplot as plt
import numpy as np

E_d0= -381.02645
E_d1= -374.48472
E_d2= -368.27568
E_t= -388.42090

mu_Co = E_Co + D_mu_Co
vmb = 6.913
mu_e = np.arange(0.0,1.3,0.01)

n0=0
n1=-1
n2=-2

E_f0= E_d0 - E_t + mu_Co + n0*(vmb + mu_e) + 0
E_f1= E_d1 - E_t + mu_Co + n1*(vmb + mu_e) + 0.73111 #0.503
E_f2= E_d2 - E_t + mu_Co + n2*(vmb + mu_e) + 1.89247 #1.233

fig=plt.figure(figsize=(9, 6))
ax=fig.add_subplot(111, label="1")
ax.tick_params(axis = 'both', which = 'major', labelsize = 22)
plt.subplots_adjust(left=0.14, bottom=0.15, right=0.95, top=0.9, wspace=00, hspace=00)
#E_f0 = E_d - E_t + mu_Co + t
plt.plot(mu_e, E_f0, label='$V_{Co}^{0}$',color="black", linewidth=3)
plt.plot(mu_e, E_f1, label='$V_{Co}^{1-}$',color="blue", linewidth=3)
plt.plot(mu_e, E_f2, label='$V_{Co}^{2-}$',color="red", linewidth=3)
M1= np.minimum(E_f0,E_f1)
M_CO=np.minimum(M1,E_f2)
plt.plot(mu_e, np.minimum(M1,E_f2),'--', linewidth=5, color='green')
plt.legend(loc="upper right", shadow=True, fontsize='x-large')
plt.legend(fontsize=25)
#plt.arrow(0.6,2.1,0.6,2.2, color="black", head_length = 0.1, head_width = 0.1)
low_x0=mu_e[8]
low_x1=mu_e[10]
low_y0=E_f1[8]
low_y1=E_f1[10]

high_x0=mu_e[8]
high_x1=mu_e[10]
high_y0=E_f2[8]
high_y1=E_f2[10]

## low_O
a_low= low_y1 - low_y0
b_low= -(low_x1 - low_x0)
c_low= a_low*low_x1 + b_low*low_y1

## high_O
a_high= high_y1 - high_y0
b_high= -(high_x1 - high_x0)
c_high= a_high*high_x1 + b_high*high_y1

#### low & Ni2O3
A=[[a_low,b_low],[a_high,b_high]]
B=[c_low,c_high]
p1_low=np.linalg.inv(A).dot(B)
print(p1_low)

plt.xlim(0,1.2)
plt.xlabel('$E_F$ (eV)', size = 22)
plt.ylabel('Formation energy (eV)', size = 22)
plt.title('$CoFe_2O_4- V_{Co}^{q}$', size = 24)
plt.grid(True)
#plt.savefig("CFO_Co.png")
plt.show()
```
